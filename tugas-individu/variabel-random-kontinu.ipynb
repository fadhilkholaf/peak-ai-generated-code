sebuah ekperimen yang mengikuti fungsi distribusi probabilitas sebagai berikut :

$$
f(x) =
\begin{cases}
\dfrac{x^2}{3}, & -1 < x < 2, \\
0, & \text{elsewhere}.
\end{cases}
$$

Buat kode didalam jupyter notebook google colab dengan urutan sebagai berikut :

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
# Definisikan fungsi PDF
def pdf(x):
    if -1 < x < 2:
        return x**2 / 3
    else:
        return 0
```

# Praktikum 5

```python
# Vektorisasi PDF untuk dapat digunakan pada input bertipe array
pdf_vec = np.vectorize(pdf)
```

```python
# Definisikan fungsi CDF dengan melakukan integrasi numerik
def cdf(x):
    result, _ = quad(pdf, -1, x)
    return result
```

```python
# Vektorisasi CDF untuk dapat digunakan pada input bertipe array
cdf_vec = np.vectorize(cdf)
```

```python
# Gambarkan PDF
x_vals = np.linspace(-1.5, 2.5, 500)
pdf_vals = pdf_vec(x_vals)
```

# Praktikum 6

```python
plt.figure(figsize=(10, 5))
plt.plot(x_vals, pdf_vals, label="PDF")
plt.fill_between(x_vals, pdf_vals, alpha=0.3)
plt.title("Probability Density Function (PDF)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Gambarkan CDF
cdf_vals = cdf_vec(x_vals)
plt.figure(figsize=(10, 5))
plt.plot(x_vals, cdf_vals, label="CDF", color="green")
plt.title("Cumulative Distribution Function (CDF)")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid(True)
plt.legend()
plt.show()
```

# Praktikum 7

```python
# Hitung probabilitas P(0 < X < 1)
(P_0_to_1, _) = quad(pdf, 0, 1)
print(f"P(0 < X < 1) = {P_0_to_1}")
```

$$
P(0 < X \leq 1) 
= \int_{0}^{1} \frac{x^2}{3}\, dx 
= \frac{1}{3} \int_{0}^{1} x^2 \, dx 
= \frac{1}{3} \left[ \frac{x^3}{3} \right]_{0}^{1} 
= \frac{1}{3} \times \left(\frac{1^3}{3} - \frac{0^3}{3}\right) 
= \frac{1}{3} \times \frac{1}{3} 
= \frac{1}{9}
$$
