Dalam soal disebutkan bahwa pada pengiriman laptop sebanyak 20 buah, didapatkan ada 3 cacat, kemudian sekolah membeli 2 buah laptop, maka kemungkinan jumlah laptop yang dibeli cacat dapat ditulis dalam variabel random X = { 0,1,2}

```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb
```

```python
# Fungsi untuk menghitung probabilitas jumlah cacat pembelian laptop
def laptop_pmf(k, N, K, n):
    """
    Menghitung probabilitas P(X = k)
    k : jumlah item cacat dalam sampel
    N : total item (populasi)
    K : total item cacat dalam populasi
    n : ukuran sampel
    """
    return (comb(K, k) * comb(N - K, n - k)) / comb(N, n)
```

# Praktikum 2

```python
# Parameter distribusi
N = 20  # Total laptop
K = 3  # Laptop cacat
n = 2  # Laptop yang dibeli
# Nilai k yang mungkin (0 cacat, 1 cacat, atau 2 cacat)
k_vals = np.arange(0, min(K, n) + 1)  # Nilai k bisa 0, 1, atau 2
# Menghitung PMF untuk setiap nilai k
pmf_vals = [laptop_pmf(k, N, K, n) for k in k_vals]
```

```python
# Gambarkan PDF
plt.figure(figsize=(10, 5))
plt.stem(k_vals, pmf_vals)
plt.title("Probability Distribution Function (PDF)")
plt.xlabel("Banyaknya Laptop Rusak")
plt.ylabel("Probability")
plt.xticks(k_vals)
plt.grid(True)
plt.show()
```

# Praktikum 3

```python
# Menghitung CDF dengan menjumlahkan PMF
cdf_vals = np.cumsum(pmf_vals)
# Gambarkan CDF
plt.figure(figsize=(10, 5))
plt.step(k_vals, cdf_vals, where="mid", label="CDF", color="green")
plt.title("Cumulative Distribution Function (CDF)")
plt.xlabel("Banyaknya Laptop Rusak")
plt.ylabel("Cumulative Probability")
plt.xticks(k_vals)
plt.grid(True)
plt.show()
```

```python
# Tampilkan tabel probabilitas
for k, pmf in zip(k_vals, pmf_vals):
    print(f"P(X = {k}) = {pmf:.4f}")
```
